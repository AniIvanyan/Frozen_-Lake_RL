```python
import gym
from gym.envs.toy_text.frozen_lake import generate_random_map
import numpy as np  

from utils import first_visit_MC
from utils import iterative_policy_evaluation
from utils import visualize_valueFuntion
```

```python
# Creates a FrozenLake environment with a random map and slippery ice.

#env=gym.make('FrozenLake-v1', desc=generate_random_map(size=4), is_slippery=True, render_mode = 'human')
env=gym.make('FrozenLake-v1', desc=generate_random_map(size=4), is_slippery=True)
```

```python
# Learns value function using first-visit MC, 1000 eps, 0.9 discount rate
estimatedValuesMonteCarlo=first_visit_MC(env,stateNumber=env.observation_space.n,numberOfEpisodes=1000,discountRate=0.9)
```

```python
# Initializes policy to be uniform
num_states = env.observation_space.n
num_actions = env.action_space.n
initial_policy = np.ones((num_states, num_actions)) / num_actions
valueFunctionVectorInitial=np.zeros(env.observation_space.n)
```

```python
# Learns value function using iterative policy evaluation, 1000 iterations, 1e-6 tolerance
estiamtedValueIterative=iterative_policy_evaluation(env,valueFunctionVectorInitial,initial_policy,1,1000,10**(-6))
```

```python
visualize_valueFuntion(estiamtedValueIterative, shape=4,fileNameToSave='iterativePolicyEvaluationEstimated.png')
```

```python
visualize_valueFuntion(estimatedValuesMonteCarlo, shape=4,fileNameToSave='MonteCarloEstimated.png')
```

```python
env.close()
```

The conclusion that can be drawn from the fact that the Monte Carlo policy evaluation and iterative policy evaluation methods give almost similar results is that this means that we can successfully teach the FrozenLake environment to solve the task of reaching the goal state by using Monte Carlo policy evaluation.
